# matplotlib
### matplotlib.org gallery for demo figures

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
```

```python
y = x**2
```

```python
y
```

```python
#FUNCTIONAL
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.show()
```

```python
# creates subplot with 1 row and 2 columns, take up 1st column
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
fig = plt.figure()
# offset x,y then size x,y 
axes1 = fig.add_axes([0.0,0.0,1,1])
# shifts right 0.1 and shift up 0.2, width is 0.5, length full
axes2 = fig.add_axes([0.1,0.2,0.5,1])
```

```python
fig = plt.figure()
# offset x,y then size x,y 
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
# shifts right 0.1 and shift up 0.2, width is 0.5, length full
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes1.set_title("LARGE")
axes2.plot(y,x)
axes2.set_title("SMALL")
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('First Plot')
axes[1].plot(y,x)
axes[1].set_title('Second Plot')
# can also use a for loop to iterate through
# plt.tight_layout <- gets rid of overlaps
```

# Figure Size and DPI

```python
fig = plt.figure(figsize=(8,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
# figure size 8 inches by 2 inches
fig,axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig('my_picture.png',dpi=200)
```

```python
fig = plt.figure(figsize=(5,5))
ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2, label='X Squared')
ax.plot(x, x**3, label='X Cubed')
# choose location 0 is best, 10 is center, etc..
ax.legend(loc=0)
# can also set using percentage like axes
ax.legend(loc=(0.1,0.1))
```

# Plot Appearance

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='orange', linewidth=10)
# or hex code
# ax.plot(x,y, color='#FF8C00')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
# alpha controls opacity
ax.plot(x,y,color='orange', linewidth=10, alpha=0.1)
ax.plot(x,y**3,color='red')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
# linestyle controls line 
ax.plot(x,y,color='orange', linewidth=10, linestyle='--')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
# marker contro
ax.plot(x,y,color='orange', linewidth=1, marker='o', markersize=20, markerfacecolor='yellow',
        markeredgewidth=2, markeredgecolor='green')
```
